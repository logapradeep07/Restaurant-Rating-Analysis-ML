```python
import pandas as pd

df = pd.read_csv("../data/Dataset .csv")
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df['restaurant_name_length'] = df['restaurant_name'].astype(str).apply(len)
df['address_length'] = df['address'].astype(str).apply(len)
df['cuisine_count'] = df['cuisines'].astype(str).apply(lambda x: len(x.split(',')))
df['has_table_booking'] = df['has_table_booking'].astype(str).str.lower().map({'yes':1, 'no':0})
df['has_online_delivery'] = df['has_online_delivery'].astype(str).str.lower().map({'yes':1, 'no':0})
df.to_csv("../data/feature_engineered_data.csv", index=False)
df.head()
```
